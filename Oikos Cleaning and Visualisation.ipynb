# Oikos Data Cleaning and Visualisation
<br>This notebook cleans the Oikos data file "REFP10_v2_graduaçao2_Ago2022_for Correlaid" and provides data analytics.
<br>

```python
#Import packages
import pandas as pd
import geopandas 
from googletrans import Translator, constants
from pprint import pprint
```

```python
#Import data
data = pd.read_excel(r"C:\Users\Barna\Desktop\Oikos Data\REFP10_v2_graduaçao2_Ago2022_for Correlaid.xlsx")
```

```python
#Translate columns to English
from deep_translator import GoogleTranslator
columns = list(data.columns)
translated_columns = []
for i in columns:
    translated = GoogleTranslator(source = "pt", target='en').translate(i)
    translated_columns.append(translated)
```

```python
#Replace column titles
data.columns = translated_columns
```

```python
#Remove photo columns
no_photo_columns = [ x for x in translated_columns if "photo" not in x ]
no_photo_columns2 = [ x for x in no_photo_columns if "Photo" not in x ]
data = data[no_photo_columns2]
```

```python
print(len(data.columns))
print(data.columns)
```

```python
#Drop completely empty columns
data_clean = data.dropna(axis=1, how='all')
# print(data_clean)
# print(len(data_clean.columns))
```

```python
#Delete personal information present
data = data_clean.drop(columns=["Beneficiary's name"])
```

```python
# Remove uninformative columns
not_needed = ["_index", "_status"]
data = data.drop(columns = not_needed)
```
